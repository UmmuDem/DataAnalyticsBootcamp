```python
import nltk
import docx2txt
from nltk.tokenize import word_tokenize
from nltk.text import Text
import pandas as pd
import tqdm
```

```python
df = pd.read_csv('to_tableau.csv')
```

```python
df.drop(columns = ['Unnamed: 0','Unnamed: 0.1','syllabus','language'])
```

```python
def get_most_common(text):
    tokens = word_tokenize(text)
    freq = nltk.FreqDist(tokens)
    word_freq = dict((word,freq) for word, freq in freq.items() if not word.isdigit())
    sorted_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse = True)
    return sorted_freq
get_most_common(df.loc[df.name == 'machine learning', 'corpus'].values[0])
```

```python
keyword_list = []
for i in tqdm.tqdm(list(df['name'])):
    keyword_list.append(get_most_common(df.loc[df.name == i , 'corpus'].values[0]))
```

```python
name = []
keyword = []
value = []
category = []
for i in range(len(keyword_list)):
    for j in range(33):
        name.append(df['name'][i])
        keyword.append(keyword_list[i][j][0])
        value.append(keyword_list[i][j][1])
        category.append(df['category'][i])

df_tableau = pd.DataFrame({'name':name,'keyword':keyword,'value':value,'category':category})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
count_vectorizer = CountVectorizer()
```

```python
def locate_corpus_by_name(user_input,df):
    user_input = user_input.lower()
    df['name']= df['name'].str.lower()
    new = df.loc[df.name == user_input, 'corpus'].values[0]
    return new
```

```python
def get_cos_similarity_scores(user_input,vectorizer,vectorized_corpus):
    tokens = [str(tok) for tok in word_tokenize(user_input)]
    user_tfidf = vectorizer.transform(tokens)
    cos_similarity = map(lambda x: cosine_similarity(user_tfidf, x),vectorized_corpus)
    output = list(cos_similarity)
    return output
```

```python
CV = count_vectorizer.fit_transform((df['corpus']))

def get_recommendation_CV(user_input,corpus,df):
    output = get_cos_similarity_scores(corpus,vectorizer = count_vectorizer, vectorized_corpus= CV)
    top = sorted(range(len(output)), key=lambda i: output[i][0], reverse=True)[:10]
    list_scores = [output[i][0][0] for i in top]
    
    recommendation = pd.DataFrame(columns = ['name','recommended_course',  'instructor','rating', 'score','index','level','about'])
    count = 0
    best_index = list()
    for i in top:
        recommendation.at[count, 'name'] = user_input
        recommendation.at[count, 'recommended_course'] = df['name'][i]
        recommendation.at[count, 'instructor'] = df['instructor'][i]
        recommendation.at[count, 'rating'] = df['rating'][i]
        recommendation.at[count, 'score'] =  list_scores[count]
        index = list(df. index)
        recommendation['index'][count] = i
        recommendation.at[count, 'about'] = df['about'][i]
        recommendation.at[count, 'level'] = df['level'][i]
        count += 1
        
        best_index.append(i) 
    return recommendation, best_index
```

```python
recommended_course = pd.DataFrame()
for i in tqdm.tqdm(list(df['name'])):
    user_input = i
    #print(user_input)
    corpus = locate_corpus_by_name(i,df)
    #print(corpus)
    collection = get_recommendation_CV(i,corpus,df)[0]
    #print(type(collection))
    recommended_course = recommended_course.append(collection)
```
