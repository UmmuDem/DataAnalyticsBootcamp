offset - allows you to retrive based on a starting position
eg offset = 201 (results 201-300)

next - while there is a next then there are more results to retrieve 

function - while next is present then run sp.next 

```python
# function which handles more than 100 results
def get_playlist_tracks(user_id,playlist_id):
    results = sp.user_playlist_tracks(user_id,playlist_id)
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
results2 = get_playlist_tracks('maka_97' ,'6mtYuOxzl58vSGnEDtZ9uB')
```
